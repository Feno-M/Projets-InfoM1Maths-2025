```python
#Projet exp_math
#exercice 2
#Constantes et algorithme PSLQ
#1)Evaluation de I
print("1)")
from mpmath import mp,quad,log ,sin ,pi
#Calculons notre integrale
def eval_I(prec):
    mp.dps=prec
    I= quad(lambda x: log(sin(x)),[0,pi/2])
    return I
I_50= eval_I(50)
I_100=eval_I(100)
I_200=eval_I(200)
print("I pour 50 decimal:",I_50)
print("I pour 100 decimal:",I_100)
print("I pour 200 decimal:",I_200)
print("2)")
mp.dps = 100
from mpmath import quad,zeta,ln,pslq,sin
I = quad(lambda x: ln(sin(x)), [0, mp.pi/2])
zeta_2 = zeta(2)  # ζ(2) = π^2/6
ln2 = ln(2)
ln2_squared = (ln2)**2
# Vecteur pour PSLQ
vector = [I, zeta_2, ln2_squared, ln2]
# Appliquer PSLQ
coefficient = pslq(vector)
if coefficient is None:
    print("Aucune relation rationnel pour ce vecteur")
else:
    print("Coefficients trouvés par PSLQ :", coefficient)
print("3)") #on'a pas trouver une relation rationnet pour vector alors on a choici vect et on a pu detecter une relation
from mpmath import quad,zeta,ln,pslq,sin,pi
I = quad(lambda x: ln(sin(x)), [0, mp.pi/2])
zeta_2 = zeta(2)  # ζ(2) = π^2/6
ln2 = ln(2)
ln2_squared = (ln2)**2
# Vecteur pour PSLQ
vect = [I, zeta_2, ln2_squared, pi*ln2]
# Appliquer PSLQ
coeff = pslq(vect)
if coeff is None:
    print("Aucune relation rationnel pour ce vecteur")
else:
    print("Coefficients trouvés par PSLQ :", coeff)
print("verification:")
from mpmath import fsum
res = fsum([a*v for a, v in zip(coeff, vect)])
print("On a bien 2*I+0*zeta(2)+0*ln(2*2)+pi*ln2=", res)
print("4)")
precisions = [30, 50, 100, 150, 200]
for p in precisions:
    mp.dps = p
    I = eval_I(p)
    vect = [I, zeta_2, ln2_squared, pi*ln2]
    rel = pslq(vect)
    print(f"Précision {p} -> Relation PSLQ : {rel}")
print("5)conjocture \n d'apres 3) on a 2I+pi*ln2=0\nalors on peut tirer que I=-pi*ln2/2 ")
print("\n\n\n\n\n\n\n\n visualisation de notre fonction")
print("\n\n")
plot(lambda x: log(sin(x)), (x, 0.01, pi/2), ymin=-10, ymax=0, title="log(sin(x)) sur [0, π/2]")
```
